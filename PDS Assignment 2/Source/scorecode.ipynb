```python
import pandas as pd
import re
from datetime import datetime
```

```python
raw_data= pd.read_csv("C:\\Users\\padmj\\OneDrive\\Documents\\PDS Assignment 2\\Raw Data\\Rawdata.csv")
```

```python
raw_data.head()
```

```python
#a)checking for null values in data frame
raw_data.isna().sum()
```

```python
df = pd.DataFrame(raw_data)
#b)mileage, engine, power and new price,has units, first we will remove these and after that we will convert into float
```

```python
df.head()
```

```python
df['Mileage'] = df['Mileage'].str.extract('(\d+\.\d+)').astype(float)

df['Engine'] = df['Engine'].str.replace(' CC', '').astype(float)

df['Power'] = df['Power'].str.extract('(\d+\.\d+)').astype(float)

df['New_Price'] = df['New_Price'].str.extract('(\d+\.\d+)').astype(float)
```

```python
#Replace the NA values of Mileage, Engine and Power with the mean
```

```python
df['Mileage'].fillna(df['Mileage'].median(),inplace=True)
df['Engine'].fillna(df['Engine'].median(),inplace=True)
df['Power'].fillna(df['Power'].median(),inplace=True)

df['Seats'].fillna(df['Seats'].mode()[0],inplace=True)

df.drop('New_Price', axis=1, inplace=True)
```

```python
df.head()
```

```python
output_csv_file = 'C:\\Users\\padmj\\OneDrive\\Documents\\PDS Assignment 2\\Results\\cleaned_data.csv'
df.to_csv(output_csv_file, index=False)
```
